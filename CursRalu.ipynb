## Cum cercetam?

  

1. cautam resurse bibliografice in numar cat mai mare in MANDALAY sau WEBOFKNOWLEDGE si luam notite sau facem fisier cu idei

2. preluam idei din ceea ce am studiat pentru a lucra pe ele

3. corelam ideile sau adaugam

4. formulam teorii reinterpretate sau noi

5. validam teoria

6. redactam
