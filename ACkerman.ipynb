# Ackerman model

The ackerman steering model is a model givivng us the equations followed by the wheels of a vehicle when having a rotational movement, taking a corner. This model will be applied ona four wheeled car. Each wheel will have it's own degree of freedom relative to the chasis. The input of the system will be considered as the desired path of the center of gravity of the Chassey.

## Modelling a vehicle

First let's define the movement of a car. The position of a car can be defined as a point in a 2-d plane, thus the position of the center of gravity of our vehicle at time $t$ will be defined as:\
$P:(X{(t)},Y{(t)}) <=> ${$X{(t)}$ is the abscise of the vehicle center relative to a fixed point in space with respect to time and $Y{(t)}$ is the ordonate of the vehicle center relative to a fixed point in space with respect to time}
Now when faced with liniar movement at any point the current position can be easily represented as:
$P(t) = \begin{cases}
  X(t) = X(0) + \int_{t_{0}}^{t} {V(t)}_{X} dt\\
  Y(t) = Y(0) + \int_{t_{0}}^{t} {V(t)}_{Y} dt
  \end{cases}$ \
where ${V(t)}_{Y}$ is the instant velocity at time $t$ with respect to the Y axis of the vehicle and ${V(t)}_{X}$ is the instant velocity at time $t$ with respect to the X axis. Assuming we start from origin we can define the current position of our vehicle at time $t$ as: \
$P(t) = \begin{cases}
  X(t) = \int_{t_{0}}^{t} {V(t)}_{X} dt\\
  Y(t) = \int_{t_{0}}^{t} {V(t)}_{Y} dt
  \end{cases}$\
Here we can see that in order to have full control of our vehicles position we need full control of our instant velocity vector.
